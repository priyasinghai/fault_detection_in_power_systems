```python
import pandas as pd
import numpy as np


df = pd.read_csv("verilog_ready_dataset.csv")

signals = ['Va', 'Vb', 'Vc', 'Ia', 'Ib', 'Ic']
samples_per_cycle = 200

feature_rows = []

for idx, row in df.iterrows():
    features = {}

    for sig in signals:

        sig_samples = np.array(
            [row[f"{sig}_{i}"] for i in range(samples_per_cycle)]
        )

        abs_sig = np.abs(sig_samples)


        features[f"{sig}_peak"] = abs_sig.max()


        features[f"{sig}_rms2"] = np.mean(sig_samples ** 2)

    features["Fault"] = row["Fault"]
    feature_rows.append(features)

feature_df = pd.DataFrame(feature_rows)


feature_df.to_csv("feature_dataset.csv", index=False)

print("Feature dataset saved as feature_dataset.csv")
```
