```python
import pandas as pd
import joblib
from sklearn.tree import DecisionTreeClassifier, export_text


df = pd.read_csv("feature_dataset.csv")

X = df.drop(columns=["Fault"])
y = df["Fault"]


dt = DecisionTreeClassifier(
    criterion="gini",
    max_depth=3,
    min_samples_leaf=5,
    random_state=42
)

dt.fit(X, y)


joblib.dump(dt, "frozen_decision_tree.pkl")


rules = export_text(dt, feature_names=list(X.columns))

with open("decision_tree_rules.txt", "w") as f:
    f.write(rules)

print("Decision tree frozen and saved.")
print("\nDecision Tree Rules:\n")
print(rules)
```
